String indexing and slicing

```python
my_string='hello'
print(my_string[0])
print(my_string[1])
print(my_string[2])
print(my_string[-2])
```

slicing

```python
print(my_string[0:3:2])
print(my_string[1:4:2])
```

slicing negative

```python
print(my_string[-1:-2:-1])
```
